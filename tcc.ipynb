```python
from statsmodels.tsa.stattools import adfuller
```

```python
print('Teste Dickey-Fuller - Arrecadação')
df_arrec_test = adfuller(df_arrec['VALOR (R$)'],autolag='AIC')
df_arrec_test
```

```python
help(adfuller)
```

```python
# Dataset Arrecadação

print('Teste Dickey-Fuller - Arrecadação')

dfout = pd.Series(df_arrec_test[0:4],index=['ADF test statistic','p-value','# lags used','# observations'])

for key,val in df_arrec_test[4].items():
    dfout[f'critical value ({key})']=val
print(dfout)
```

Aqui temos um valor-p muito alto em 0,998, o que fornece evidências fracas contra a hipótese nula e, portanto, falhamos em rejeitar a hipótese nula e decidimos que nosso conjunto de dados não é estacionário.

```python
print('Teste Dickey-Fuller - Saques')
df_saque_test = adfuller(df_saque['VALOR (R$)'],autolag='AIC')
df_saque_test
```

```python
# Dataset de Saques

print('Teste Dickey-Fuller - Saques')

dfout = pd.Series(df_saque_test[0:4],index=['ADF test statistic','p-value','# lags used','# observations'])

for key,val in df_saque_test[4].items():
    dfout[f'critical value ({key})']=val
print(dfout)
```

## Avaliação de Desempenho

```python
from sklearn.metrics import mean_squared_error
from statsmodels.tools.eval_measures import rmse
```

### Arrecadação

```python
df = df_arrec.iloc[11:-1]
```

```python
mse_sma = mean_squared_error(df['VALOR (R$)'], df['SMA12'])
mse_ewma = mean_squared_error(df['VALOR (R$)'], df['EWMA12'])
mse_des = mean_squared_error(df['VALOR (R$)'], df['DES'])
mse_tes = mean_squared_error(df['VALOR (R$)'], df['TES'])


rmse_sma = rmse(df['VALOR (R$)'], df['SMA12'])
rmse_ewma = rmse(df['VALOR (R$)'], df['EWMA12'])
rmse_des = rmse(df['VALOR (R$)'], df['DES'])
rmse_tes = rmse(df['VALOR (R$)'], df['TES'])

print('-------------------------------\nMSE (Mean Squared Error):\n')
print(f'MSE SMA:  {mse_sma/10**12:11.8} tri')
print(f'MSE EWMA: {mse_ewma/10**12:11.8} tri')
print(f'MSE  DES: {mse_des/10**12:11.6} tri')
print(f'MSE TES:  {mse_tes/10**12:11.8} tri')
print('-------------------------------\nRMSE (Root Mean Squared Error):\n')
print(f'RMSE  SMA: {rmse_sma/10**6:11.4} mi')
print(f'RMSE EWMA: {rmse_ewma/10**6:11.4} mi')
print(f'RMSE DES:  {rmse_des/10**6:11.3} mi')
print(f'RMSE TES:  {rmse_tes/10**6:11.4} mi')
```

# Modelo AR

```python
from statsmodels.tsa.ar_model import AutoReg,ARResults
```

```python
train_data_AR = df_arrec['VALOR (R$)'].iloc[:286]
test_data_AR = df_arrec['VALOR (R$)'].iloc[286:]
```

```python
ARfit = AutoReg(train_data_AR, lags = 12).fit()
print(f'Lag: {ARfit.ar_lags}')
print(f'Coefficients:\n{ARfit.params}')
```

```python
# This is the general format for obtaining predictions
start=len(train_data_AR)
end=len(train_data_AR)+len(test_data_AR)-1
predictions1 = ARfit.predict(start=start, end=end, dynamic=False).rename('AR(1) Predictions')
```

```python
test_data_AR.plot(legend=True)
predictions1.plot(legend=True,figsize=(12,6));
```

```python
mean_absolute_error(test_data_AR,predictions1)
```

```python
mean_squared_error(test_data_AR,predictions1)
```

```python
np.sqrt(mean_squared_error(test_data_AR,predictions1))
```
